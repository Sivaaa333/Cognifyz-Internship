```python
def get_recommendations(restaurant_index, cosine_sim=cosine_sim):
    sim_scores = list(enumerate(cosine_sim[restaurant_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    restaurant_indices = [index for index, _ in sim_scores]
    return df['Restaurant Name'].iloc[restaurant_indices]
```
