```python
# Assuming x_test contains the test features
x_test_encoded = pd.get_dummies(x_test, columns=['Locality Verbose', 'Has Table booking', 'Has Online delivery',
       'Is delivering now', 'Switch to order menu'])
```

```python
# Check for any columns present in x_test but not in x_train_encoded
missing_columns = set(x_train_encoded.columns) - set(x_test_encoded.columns)
if missing_columns:
    # Handle missing columns in x_test_encoded (e.g., set them to zero)
    x_test_encoded[list(missing_columns)] = 0
```

```python
# Check for any columns present in x_test_encoded but not in x_train_encoded
new_columns = set(x_test_encoded.columns) - set(x_train_encoded.columns)

if new_columns:
    # Handle new columns in x_test_encoded (e.g., set them to zero)
    x_test_encoded[list(new_columns)] = 0

```

```python
x_test_encoded = x_test_encoded[x_train_encoded.columns]
```

```python
y_pred = regression_model.predict(x_test_encoded)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

```
